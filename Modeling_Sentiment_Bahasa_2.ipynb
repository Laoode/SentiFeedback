```python
# Menghapus baris dengan missing values
data = data.dropna()

# Memastikan tidak ada missing values lagi
print(data.isnull().sum())

# Melihat jumlah baris setelah penghapusan
print(f"Jumlah baris setelah penghapusan: {data.shape[0]}")
```

```python
# Melihat nilai unik pada kolom 'Rating' dan jumlah kemunculannya
rating_counts = data['Rating'].value_counts()

# Menampilkan hasil
print(rating_counts)
```

```python
# Definisikan mapping pertama
map_initial = {
    'awesome': 'Sangat Baik',
    'good': 'Baik',
    'average': 'Cukup Baik',
    'poor': 'Buruk',
    'awful': 'Sangat Buruk'
}

# Terapkan mapping pertama
data['Rating'] = data['Rating'].replace(map_initial)

# Definisikan mapping kedua untuk menyederhanakan kategori
map_simplified = {
    'Sangat Baik': 'Baik',
    'Sangat Buruk': 'Buruk',
    'Cukup Baik': 'Baik'
}

# Terapkan mapping kedua
data['Rating'] = data['Rating'].replace(map_simplified)

# Menampilkan distribusi nilai setelah mapping
print(data['Rating'].value_counts())
```

```python
import re

# Menghapus kolom 'Id'
data_cleaned = data.drop(columns=['Id'])

# Menghapus punctuation, emoji, dan casefolding
def clean_text(text):
    # Hapus emoji memakai regec
    text = re.sub(r'[^\w\s]', '', text)  # Menghapus punctuation
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # Menghapus karakter non-ASCII(emoji)
    text = text.lower()  # Konversi ke lowercase
    return text

# Apply cleaning ke kolom 'Komentar'
data_cleaned['Komentar_bersih'] = data_cleaned['Komentar'].apply(clean_text)

# Preview
data_cleaned.head()
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
import pandas as pd

# Inisialisasi StopWordRemover
factory = StopWordRemoverFactory()
stopword_remover = factory.create_stop_word_remover()

# Fungsi untuk menghapus stopwords
def remove_stopwords(text):
    return stopword_remover.remove(text)

# Terapkan stopword removal pada kolom 'Komentar_bersih_tokenized' (bukan 'Komentar_bersih' setelah tokenisasi)
data_cleaned['Komentar_bersih_stopwords'] = data_cleaned['Komentar_bersih_tokenized'].apply(lambda x: ' '.join([remove_stopwords(word) for word in x]))

data_cleaned[['Komentar_bersih_tokenized', 'Komentar_bersih_stopwords']].head()
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Inisialisasi Stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming
def apply_stemming(text):
    return stemmer.stem(text)

# Terapkan stemming pada kolom 'Komentar_bersih_stopwords'
data_cleaned['Komentar_bersih_stemmed'] = data_cleaned['Komentar_bersih_stopwords'].apply(apply_stemming)

# Lihat hasil stemming
print(data_cleaned[['Komentar_bersih_stopwords', 'Komentar_bersih_stemmed']].head())
```

```python
# Map nilai pada kolom 'Rating' ke angka
rating_to_numeric = {'Buruk': 1, 'Baik': 0}

# Terapkan mapping ke kolom 'Rating'
data_cleaned['Rating_num'] = data_cleaned['Rating'].map(rating_to_numeric)

data_cleaned
```

# Modeling

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import pandas as pd

# Menggunakan TfidfVectorizer untuk teks
tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # Membatasi jumlah fitur (kata)
X_tfidf = tfidf_vectorizer.fit_transform(data_selected['Komentar_bersih_stemmed'])

# Konversi hasil ke dalam bentuk array atau DataFrame
X_tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Target variabel (Rating_num)
y = data_selected['Rating_num'].values

# Membagi data menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X_tfidf_df, y, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import numpy as np

# Menambahkan dimensi ekstra agar cocok untuk LSTM (batch_size, sequence_length, features)
X_train_lstm = np.expand_dims(X_train.values, axis=-1)  # Membuat dimensi menjadi (samples, features, 1)
X_test_lstm = np.expand_dims(X_test.values, axis=-1)    # Membuat dimensi menjadi (samples, features, 1)

# Membangun model LSTM
model = Sequential()

# Menambahkan layer LSTM
model.add(LSTM(64, input_shape=(X_train_lstm.shape[1], 1)))  # Sesuaikan dengan bentuk input

# Menambahkan dropout untuk mencegah overfitting
model.add(Dropout(0.5))

# Layer output
model.add(Dense(1, activation='sigmoid'))  # Aktivasi sigmoid untuk klasifikasi biner

# Kompilasi model
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Menampilkan ringkasan model
model.summary()
```

```python
# Melatih model LSTM
history = model.fit(X_train_lstm, y_train, epochs=5, batch_size=64, validation_data=(X_test_lstm, y_test))
```

```python
# Evaluasi akurasi pada data training
train_loss, train_accuracy = model.evaluate(X_train_lstm, y_train)
print(f"Training Loss: {train_loss}")
print(f"Training Accuracy: {train_accuracy}")

# Evaluasi akurasi pada data testing
test_loss, test_accuracy = model.evaluate(X_test_lstm, y_test)
print(f"Testing Loss: {test_loss}")
print(f"Testing Accuracy: {test_accuracy}")
```

```python
# Kalimat baru yang ingin diprediksi
new_comment = "Buruk"
# new_comment = "Baik"


new_comment_tfidf = tfidf_vectorizer.transform([new_comment])

# Konversi ke array dan tambahkan dimensi ekstra untuk input ke model LSTM
new_comment_lstm = np.expand_dims(new_comment_tfidf.toarray(), axis=-1)

# Prediksi
predicted_prob = model.predict(new_comment_lstm)[0][0]
predicted_label = "Baik" if predicted_prob >= 0.5 else "Buruk"

# Menampilkan hasil
print(f"Kalimat: {new_comment}")
print(f"Probabilitas Baik: {predicted_prob:.2f}")
print(f"Prediksi: {predicted_label}")
```
