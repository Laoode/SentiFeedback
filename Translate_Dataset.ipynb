```python
import pandas as pd
```

```python
df = pd.read_csv('/content/feedback_student.csv', encoding='latin1')
```

```python
df.head()
```

```python
df['comment'] = df['comment'].fillna("Unknown")
df['quality'] = df['quality'].fillna("Unknown")
print(df['comment'].isnull().sum())
print(df['quality'].isnull().sum())
```

```python
from googletrans import Translator
translator = Translator()

def translate_text(text):
    try:
        return translator.translate(text, dest='id').text
    except Exception as e:
        return text


df['comment_translated'] = df['comment'].apply(translate_text)
df['quality_translated'] = df['quality'].apply(translate_text)


translated_file = '/content/dataset_translated.csv'
df.to_csv(translated_file, index=False)
```

```python
df.head()
```

```python
unique_values = df['quality_translated'].unique()
print(unique_values)
```

```python
mapping = {
    'buruk sekali': 'Sangat Buruk',
    'miskin': 'Buruk',
    'rata-rata': 'Cukup Baik',
    'Bagus': 'Baik',
    'luar biasa': 'Sangat Baik'
}


df['quality_translated'] = df['quality_translated'].replace(mapping)


print(df)
```

```python
df = df.drop(columns=['comment', 'quality'])
print(df)
```

```python
df.rename(columns={
    'comment_translated': 'komentar',
    'quality_translated': 'rating'
}, inplace=True)


print(df)
```

```python
df.rename(columns={
    'komentar': 'Komentar',
    'rating': 'Rating'
}, inplace=True)

print(df)
```
